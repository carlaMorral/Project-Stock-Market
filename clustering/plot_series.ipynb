```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
comps = ['cag', 'leg', 'mo', 'nke', 'rost', 'sbux', 'see', 'wec']

with open("../data.csv", "r") as f:
    df = pd.read_csv(f)
    grouped_df = df.groupby('Stock').apply(lambda x: list(x['Price']))
    lengths = grouped_df.apply(len)
    maxlen = max(np.array(lengths.values))
    grouped_df = grouped_df[grouped_df.apply(len) == maxlen]

X = list(range(maxlen))
for comp in comps:
    Y = grouped_df[comp]
    plt.plot(X, Y, label=comp)

plt.xlabel('Days')
plt.ylabel('Prices')
plt.title('Stocks')
plt.legend()
plt.show()
```
